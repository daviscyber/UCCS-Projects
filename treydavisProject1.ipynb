```python
# Importing pandas into Python for data analysis, and matplotlib for designing charts. .

import pandas as pd
import matplotlib.pyplot as plt

gameStats = pd.read_csv('http://infs3080-class.uccs.edu/pydev/dataFiles/VIDEO-GAMES/Video_Games.csv')


# Showing the info of the dataframe in order to understand the data I am working with

gameStats.info()
```

```python
# Printing out the top 5 results to get a look at how the data is laid out.

gameStats.head()
```

```python
# Shaping the data to only show the video game names and global sales

salesFinal = gameStats[['Name', 'Global_Sales']]

# Printing out the new dataframe with just the Name and Global Sales columns.
salesFinal
```

```python
# Sorting the dataframe by the top selling global video game.

salesFinal.sort_values(by=['Global_Sales'], ascending=False)
```

```python
# The top video game by units is Wii Sports, at 82.53 million units sold.

salesFinal.head(1)
```

```python
# Shaping the data to show only the video game name and the user count

userSorted = gameStats[['Name', 'User_Count']]

# Printing out the new dataframe

userSorted
```

```python
# Sorting the dataframe to have the top game by user count as the top result.

topUser = userSorted.sort_values(by="User_Count", ascending=False)

topUser
```

```python
# The top video game by user count is The Witcher 3: Wild Hunt.

topUser.head(1)
```

```python
# Shaping the data to show top video games by units sold in North America.

naGameStats = gameStats[['Name', 'NA_Sales']]

# Creating a new dataframe that is sorted by NA sales of video games.

naSales = naGameStats.sort_values(by="NA_Sales", ascending=False)

# Printing out the top 10 results of the new dataframe.

naSales.head(10)
```

```python
# Creating a dictionary of the sales data to graph.

game_titles = ["Wii Sports","Super Mario Bros.","Duck Hunt","Tetris","Mario Kart Wii","Wii Sports Resort","Kinect Adventures!", "New Super Mario Bros. Wii","Wii Play","Super Mario World"]

na_sales = [41.36, 29.08, 26.93, 23.20, 15.68,15.61, 15.00, 14.44, 13.96, 12.78]


# Creating the bar chart.

plt.figure(figsize=(10, 4))  
plt.bar(game_titles, na_sales, color='black')

# Creating the x and y axis titles, as well as the bar chart title.

plt.xlabel('Game Title')
plt.ylabel('NA Sales (millions)')
plt.title('Top 10 Games by NA Sales')

# Rotating the x axis for better readability.

plt.xticks(rotation=45)    

# Showing the created bar chart.

plt.show()
```

```python
# Creating a new dataframe with video game name, North American sales, and the year of game release.

naDecadeSales = gameStats[['Name', 'NA_Sales', 'Year_of_Release']]

# Printing out the new dataframe to ensure the shaping of data worked. 

naDecadeSales
```

```python
# Shaping the data to show just games released during and after the year 2010.

recentDecade = naDecadeSales[naDecadeSales['Year_of_Release'] >= 2010]

# Printing out the dataframe to ensure the shaping worked.

recentDecade
```

```python
# Sorting the dataframe by NA Sales with the 2010 filter on it.

top2010 = recentDecade.sort_values(by="NA_Sales", ascending=False)

# Printing the dataframe to ensure it worked.

top2010
```

```python
# Printing out the top 10 games by North American sales, from the year 2010 and after.

top2010.head(10)
```

```python
# Creating a dictionary of the sales data to graph. 

games = ["Kinect Adventures!","Call of Duty: Black Ops","Grand Theft Auto V","Call of Duty: Modern Warfare 3","Call of Duty: Black Ops II","Halo: Reach","Grand Theft Auto V","Call of Duty: Ghosts","Halo 4","Call of Duty: Black Ops 3"]

sales = [15.00, 9.70, 9.66, 9.04, 8.25, 7.04, 7.02, 6.73, 6.65, 6.03]


# Creating the bar chart.

plt.figure(figsize=(10, 4))
plt.bar(games, na_sales, color='black')

# Creating the labels for the x and y axis, as well as the title of the chart.

plt.xlabel('Game Title')
plt.ylabel('Units')
plt.title('Top 10 NA Games by Units Since 2010')

# Rotating the x axis labels for better readability.

plt.xticks(rotation = 45)  

# Adding in horizontal gridlines for units.

plt.grid(axis='y', linestyle='--') 

# Showing the final result of the bar chart.

plt.show()
```

```python
# Grouping the data by Year of Release and summing up global sales by each year

total_sales_by_year = gameStats.groupby('Year_of_Release')['Global_Sales'].sum()

# Creatin the line chart.

plt.figure(figsize=(10, 4))
plt.plot(total_sales_by_year.index, total_sales_by_year.values, marker='o', linestyle='-')

# Labeling the x and y axis, as well as the title of the line chart.

plt.title('Total Units Sold Over the Years')
plt.xlabel('Year')
plt.ylabel('Total Units Sold')

# Implementing a grid for the line chart for better readability of data points.

plt.grid(True)

# Showing the line chart.

plt.show()
```
