```python
# Importing pandas into python for data analysis.

import pandas as pd
```

```python
# Importing the IMDB data files as dataframes with relevant names.

dfTitles = pd.read_csv('http://infs3080-class.uccs.edu/pydev/dataFiles/IMDB/titleBasics.tsv', sep='\t', low_memory=False)
dfRatings = pd.read_csv('http://infs3080-class.uccs.edu/pydev/dataFiles/IMDB/ratings.tsv', sep='\t', low_memory=False)
dfNames = pd.read_csv('http://infs3080-class.uccs.edu/pydev/dataFiles/IMDB/nameBasics.tsv', sep='\t', low_memory=False)
dfCrew = pd.read_csv('http://infs3080-class.uccs.edu/pydev/dataFiles/IMDB/crew.tsv', sep='\t', low_memory=False)
```

```python
# Printing out the info for each dataframe to see what kind of data I am working with.

dfTitles.info()
print()
dfRatings.info()
print()
dfNames.info()
print()
dfCrew.info()
```

```python
# Renaming ncsont to tconst in order to concatenate dfNames with dfRatings properly.

dfNames.rename(columns={'nconst':'tconst'}, inplace=True)

# Ensuring the rename worked properly.

dfNames.info()
```

```python
result = pd.concat([dfNames, dfRatings], axis=1)

result
```

```python
nameSort = (result.query('primaryProfession == "director"'))

nameSort
```

```python
keptColumns = ['primaryName', 'averageRating', 'numVotes']

nameSort = nameSort[keptColumns]

nameSort
```

```python
nameSort.sort_values(by=['averageRating', 'numVotes'], ascending=False)
```

```python
# This is a dataframe to sort by averageRating and numVotes from the top to the bottom.

semi1 = (nameSort.sort_values(by=['averageRating', 'numVotes'], ascending = False))

# This dataframe contains only the information we want to judge by.
semi2 = semi1[['primaryName', 'averageRating', 'numVotes']]
              
# This dataframe contains just the head of the dataframe containing the values we want to judge, or the top
# rated by rating and number of votes.
dirFinal = (semi2.head(1))
              
# Printing out the top director
dirFinal
```

```python
# Exporting the final resullt to a csv file

dirFinal.to_csv('question1.csv', index=False)
```

```python
# Concatenating dfTitles and dfRatings together to get ratings and movies in one dataframe.

mvResult = pd.concat([dfTitles, dfRatings], axis = 1)

# Printing the results.

mvResult
```

```python
# Sorting the movies by movie, excluding things like TV series and shorts.

mvSort = (mvResult.query('titleType == "movie"'))

# Sorting by the year 2000, excluding all other years.

mvSort = (mvResult.query('startYear == "2000"'))

# Print the results.

mvSort
```

```python
# Keeping only the desired columns

keptValue = ['startYear', 'primaryTitle', 'numVotes', 'averageRating']

mvSort = mvSort [keptValue]

# Printing out the result

mvSort
```

```python
# Sorting the movies by averageRating and numVotes

mvSort.sort_values(by=['averageRating', 'numVotes'], ascending=False)
```

```python
# This is a dataframe to sort by averageRating and numVotes from the top to the bottom.

semi3 = (mvSort.sort_values(by=['averageRating', 'numVotes'], ascending = False))

# This dataframe contains only the information we want to judge by.
semi4 = semi3[['startYear', 'primaryTitle', 'numVotes', 'averageRating']]
              
# This dataframe contains just the head of the dataframe containing the values we want to judge, or the top
# rated by rating and number of votes.
final = (semi4.head(5))
              
# Printing out the top 5 movies
final
```

```python
final.to_csv('question2.csv', index=False)
```
